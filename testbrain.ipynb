```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('C:/Users/Ayan Kanti Das/OneDrive/Desktop/brain/braintumor.h5')

# Explicitly compile the model
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])

# Load and preprocess the image
```

```python
import cv2
import numpy as np


img = cv2.imread('15 no.jpg')

img = cv2.resize(img, (150, 150))

img_array = np.array(img)

img_array.shape
```

```python
img_array = img_array.reshape(1,150,150,3)
img_array.shape
```

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
img = image.load_img('15 no.jpg')
plt.imshow(img, interpolation='nearest')
```

```python
ans = model.predict(img_array)
indices = ans.argmax()

if indices == 0 :
    print("Glioma Tumor Found")
elif indices == 1:
    print("Meningioma Tumor Found")
elif indices == 2:
    print("No Tumor Found")
elif indices == 3:
    print("Pituitary Tumor Found")
```
